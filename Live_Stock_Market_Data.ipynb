# **Graph Of Live Stock Market Data**

**I. Import Package**

The first step is to import the aforementioned packages. Imports the necessary packages we will need to retrieve and plot live market data. Once the required packages are imported, we can begin utilizing them to retrieve market data.

```python
#importing libraries
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import datetime as dt
import plotly.graph_objs as go
```

**II. Specify Which Stock We Wish To Graph (Ticker)**

With the yahoo finance API and pandas data reader, we can write code that will allow us to import data from any publicly traded stock symbol.

```python
#Overriding Yahoo Finance
yf.pdr_override()
```

```python
# Create input field for our desired stock
stock=input("Enter a stock symbol: ")
```

```python
# Retrieve stock data frame (df) from yfinance API at an interval of 1m
df=yf.download(tickers=stock,period='1d', interval='1m')
print(df)
```

**III. Visualizing Our Data**

In order to visualize our data we will need to declare and set up our graph. This can be done using Plotly with the following code:(Plotly is defined as go in our code)

```python
# Declare plotly figure (go)
fig=go.Figure()
fig.add_trace(go.Candlestick(x=df.index, open=df['Open'],high=df['High'], low=df['Low'], close=df['Close'], name='market data'))
fig.update_layout(title=str(stock)+ 'Live Share price:', yaxis_title='Stock Price(USD per Shares)')
fig.update_xaxes( rangeslider_visible=True, rangeselector=dict( buttons=list([dict(count=15, label="15m", step="minute", stepmode="backward"), dict(count=45, label="45m", step="minute", stepmode="backward"), dict(count=1, label="HTD", step="hour", stepmode="todate"), dict(count=3, label="3h", step="hour", stepmode="backward"), dict(step="all")])))
fig.show()
```
