# **Hierarchical Clustering**

Hierarchical clustering is a popular method for grouping objects. It creates groups so that objects within a group are similar to each other and different from objects in other groups. Clusters are visually represented in a hierarchical tree called aÂ dendrogram.
Hierarchical clustering has a couple of key benefits:


*   There is no need to pre-specify the number of clusters. Instead, the dendrogram can be cut at the appropriate level to obtain the desired number of clusters.

*  Data is easily summarized/organized into a hierarchy using dendrograms. Dendrograms make it easy to examine and interpret clusters.

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Firstly, we would create diffrent clusters/groups. To *create* diffrent clusters from  Normal distribution using *np.random.normla()*function.

In next six line of code we are creating three sets (x1,y1), (x2,y2),(x3,y3)

```python
# cluster no.1
x1=np.random.normal(25,5,100)
y1=np.random.normal(25,5,100)
```

```python
# cluster no. 2
x2=np.random.normal(55,5,100)
y2=np.random.normal(60,5,100)
```

```python
# cluster no. 3
x3=np.random.normal(55,5,100)
y3=np.random.normal(15,5,100)
```

Now, we would bring together the 3 sets of values together by concating them into a single set of values (x,y). in next lines we are doing the same.

```python
x=np.concatenate((x1,x2,x3),axis=0)
y=np.concatenate((y1,y2,y3),axis=0)
```

```python
dictonary={'x':x,'y':y}
data=pd.DataFrame(dictonary)
```

Next, we plot the values on the graph. We can see in the output that we have 3 diffrent clusters.

```python
plt.scatter(x1,y1,color='black')
plt.scatter(x2,y2,color='black')
plt.scatter(x3,y3,color='black')
```

Let's color the points andlabel the axis. Also createthe hierarchical clustering plot. For this task we would be required with following libraries:

*   linkage and dendrogram from scipy.cluster.hierarchy library
*   AgglomerativeClustering from sklearn.cluster


```python
from scipy.cluster.hierarchy import linkage, dendrogram
merg=linkage(data, method='ward')
dendrogram(merg,leaf_rotation=90)
plt.xlabel('data points')
plt.ylabel('eculidian distance')
plt.figure()
```

```python
from sklearn.cluster import AgglomerativeClustering
hiyerartical_cluster= AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
cluster=hiyerartical_cluster.fit_predict(data)
data['label']=cluster
plt.figure()
plt.scatter(data.x[data.label==0],data.y[data.label==0],color='red')
plt.scatter(data.x[data.label==1],data.y[data.label==1],color='blue')
plt.scatter(data.x[data.label==2],data.y[data.label==2],color='orange')
plt.show()
```
