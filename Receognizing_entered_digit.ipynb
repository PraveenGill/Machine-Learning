# **Machine learning model for recoganising digits of entered number.**


Our recognition process contains three steps:

1.  Train the system to guess the numbers via training data
2.  Get input digits for training
3. Test the system with new/unknown data







```python
#importing libraries
import numpy as np # linear algebra
import random
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import datasets
import matplotlib.pyplot as plt
```

Loading Training dataset

```python
# Load the digits dataset: digits

digits = datasets.load_digits()
```

Getting input from user and displying result


```python
#display 1010 digit
#taking input from user
n=input("enter a number:")
#displaying each digit using for loop
for el in n:
    el=int(el)
    plt.imshow(digits.images[el], cmap=plt.cm.plasma, interpolation='nearest')
    plt.title(digits.target[el], fontsize=20)
    plt.show()
```
